//Create Data Frame and make copy of it

```python
import pandas as pd
```

```python
raw_csv_data=pd.read_csv('Absenteeism_data.csv')
df=raw_csv_data.copy()
```

// Set size of displaying table////////

```python
pd.options.display.max_columns=None
pd.options.display.max_rows=10
```

```python
df['Reason for Absence'].unique()
```

```python
len(df['Reason for Absence'].unique())
```

//To convert categorical variables into dummies and delete nonreasonable column////////

```python
reason_columns=pd.get_dummies(df['Reason for Absence'],drop_first=True)
reason_columns
```

```python
df=df.drop(['ID'], axis=1)
```

//The sum of every row no more than 1, hence we use max function
//Reason 1-Group 1: Columns 1 to 148

//Reas - Group 2: Columns 15, 16, and 17on 2
//Reas - Group 3: Columns 18, 19, 20, and 21on 3e//REas - Group 4: Columns 22 to 28on 4

```python
reason_type_1= reason_columns.loc[:,1:14].max(axis=1)
reason_type_2= reason_columns.loc[:,5:17].max(axis=1)
reason_type_3= reason_columns.loc[:,18:21].max(axis=1)
reason_type_4= reason_columns.loc[:,22:].max(axis=1)
```

//Concatinate to df 4 columns(4 Reason types)

```python
df.columns.values
```

```python
df=pd.concat([df,reason_type_1,reason_type_2, reason_type_3, reason_type_4], axis=1)
column_names=['Reason for Absence', 'Date', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
df.columns=column_names
```

```python
df.head()
```

```python
df=df.drop(['Reason for Absence'], axis=1)
```

```python
column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 
                          'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df=df[column_names_reordered]
df.head()
```

```python
df_reason_mode=df.copy()
```

//Set type of Date column and extract month value and week value/////////////////

```python
type(df_reason_mode['Date'][0])
```

```python
df_reason_mode['Date']=pd.to_datetime(df_reason_mode['Date'], format='%d/%m/%Y')
```

```python
df_reason_mode.head()
```

```python
df_reason_mode['Date'][0].month
```

```python
month_list=[]
for i in range(df_reason_mode.shape[0]):
    month_list.append(df_reason_mode['Date'][i].month)
```

```python
df_reason_mode['Month Value']=month_list
```

```python
df_reason_mode['Date'][0].weekday()
```

```python
def date_to_week(day_value):
    return day_value.weekday()
```

```python
df_reason_mode['Day of the week']=df_reason_mode['Date'].apply(date_to_week)
```

```python
df_reason_mode=df_reason_mode.drop(['Date'],axis=1)
```

```python
column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 
                          'Month Value','Day of the week', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df_reason_mode=df_reason_mode[column_names_reordered]
df_reason_mode.head()
```

```python
df_reason_date_mod=df_reason_mode.copy()
```

```python
df_reason_date_mod['Education']=df_reason_date_mod['Education'].map({1:0,2:1,3:1,4:1})
```
